```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from google.colab.patches import cv2_imshow
```

```python
# Change the path to where you have saved the text file.
path="/content/drive/MyDrive/machineLearning/Assignment0/test.png"
img = cv2.imread(path)
```

```python
ref1 = np.array([255, 0, 0])
ref2 = np.array([0, 255, 0])
ref3 = np.array([0, 0, 255])

myarr1 = img-ref1
myarr2 = img-ref2
myarr3 = img-ref3

norm1 = np.linalg.norm(myarr1, axis=2)
norm2 = np.linalg.norm(myarr2, axis=2)
norm3 = np.linalg.norm(myarr3, axis=2)
```

```python
cv2_imshow(norm1)
cv2_imshow(norm2)
cv2_imshow(norm3)
```

```python
newImg = np.copy(img)
for i in range(img.shape[0]):
  for j in range (img.shape[1]):
    if (norm1[i,j]<norm2[i,j]):
      if (norm1[i,j]<norm3[i,j]):
        newImg[i,j] = ref1
      else:
        newImg[i,j] = ref3
    elif (norm2[i,j]<norm1[i,j]):
      if (norm2[i,j]<norm3[i,j]):
        newImg[i,j] = ref2
      else:
        newImg[i,j] = ref3

# change the path1 to the location you want to save the output1
path1="/content/drive/MyDrive/machineLearning/Assignment0/output1.png"
cv2.imwrite(path1,newImg)
cv2.imwrite('output1.png',newImg)
cv2_imshow(newImg)
```

```python
newImg2 = np.copy(img)
indW = img.shape[0]/2 - 25
indE = img.shape[0]/2 + 25
indN = img.shape[1]/2 + 25
indS = img.shape[1]/2 - 25
ref = [0,0,0]
for i in range(img.shape[0]):
  for j in range(img.shape[1]):
    if (i>= indW) and (i<=indE):
      if (j>=indS) and (j<=indN):
        newImg2[i,j] = ref

# change the path2 to the location you want to save the output2
path2="/content/drive/MyDrive/machineLearning/Assignment0/output2.png"
cv2.imwrite(path2,newImg2)
cv2.imwrite('output2.png',newImg2)
cv2_imshow(newImg2)
```
